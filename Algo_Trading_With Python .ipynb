#### Creating  our first strategy function 

```python
def test_strategy(stock, start, end, SMA):
    # downloading data 
    df = yf.download(stock, start=start, end=end)
    # grabbing Close data from dataframe 
    data = df.Close.to_frame()
    # logarithmic returns
    data["returns"] = np.log(data.Close.div(data.Close.shift(1)))
    # buy and hols fpr short or long 
    data["SMA_S"] = data.Close.rolling(int(SMA[0])).mean()
    data["SMA_L"] = data.Close.rolling(int(SMA[1])).mean()
    # dropping nan values 
    data.dropna(inplace=True)
    # defining positioning for short & long hold 
    data["position"]=np.where(data["SMA_S"] > data["SMA_L"], 1,-1)
    # defining strategy for buy and hold 
    data["strategy"]=data["returns"]*data.position.shift(1)
    # droopping nan values 
    data.dropna(inplace=True)
    # exponential returns 
    ret=np.exp(data["strategy"].sum())
    # annual standard deviation 
    std=data["strategy"].std()*np.sqrt(252)
    
    return ret, std
    
    
```

```python
# Lets create class : reason for creating class so that we can use these functions in other python file too throuugh importing rather copying the whole functions. 


class SMABacktester():
    def __init__(self, symbol, SMA_S, SMA_L, start, end):
        self.symbol = symbol 
        self.SMA_S = SMA_S 
        self.SMA_L = SMA_L 
        self.start = start
        self.end = end
        self.results = None 
        self.get_data()
        
    def get_data(self):
        # downloading data 
        df = yf.download(self.symbol, start=self.start, end=self.end)
        # grabbing Close data from dataframe 
        data = df.Close.to_frame()
        # logarithmic returns
        data["returns"] = np.log(data.Close.div(data.Close.shift(1)))
        # buy and hols fpr short or long 
        data["SMA_S"] = data.Close.rolling(self.SMA_S).mean()
        data["SMA_L"] = data.Close.rolling(self.SMA_L).mean()
        # dropping nan values 
        data.dropna(inplace=True)
        # storing data 
        self.data2 = data 
        
        
        return data
    
    def test_results(self):
        # copying data from data2 
        data = self.data2.copy().dropna()
        # defining positioning for short & long hold 
        data["position"]=np.where(data["SMA_S"] > data["SMA_L"], 1,-1)
        # defining strategy for buy and hold 
        data["strategy"]=data["returns"]*data.position.shift(1)
        # droopping nan values 
        data.dropna(inplace=True)
        # Comparison with buy and hold with respect to strategy 
        data["returnsbh"] = data["returns"].cumsum().apply(np.exp)
        data["returnsstrategy"] = data["strategy"].cumsum().apply(np.exp)
        perf = data["returnsstrategy"].iloc[-1]
        outperf = perf - data["returnsbh"].iloc[-1]
        self.results=data
        
        # exponential returns 
        ret=np.exp(data["strategy"].sum())
        # annual standard deviation 
        std=data["strategy"].std()*np.sqrt(252)

        #return ret, std
        return round(perf, 6), round(outperf, 6)
    
    def plot_results(self):
        if self.results is None:
            print("Run the test please")
        else:
            title="{}| SMA_S={} | SMA_L{}".format(self.symbol, self.SMA_S, self.SMA_L)
            self.results[["retrunsbh", "retrunsstrategy"]].plot(title=title, figsize=(12, 8))
            

        
        
        
```
